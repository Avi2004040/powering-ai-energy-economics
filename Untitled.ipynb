## Gas seasonality
Natural gas is primarily a heating fuel, so demand and price peaks in
winter and troughs in summer. This is a well-established seasonal pattern
in North American gas markets. I test it below using Henry Hub monthly spot prices (EIA, 1997–2026).

```python
storage = pd.read_excel("data/raw/NG_STOR_WKLY_S1_W.xls", sheet_name="Data 1", skiprows=2)
storage = storage.rename(columns={
    "Weekly Lower 48 States Natural Gas Working Underground Storage (Billion Cubic Feet)": "storage_bcf"})

storage = storage[["Date", "storage_bcf"]]
storage["year"] = storage["Date"].dt.year
storage["month"] = storage["Date"].dt.month

storage.head()
```

### The storage cycle
Plotting weekly storage over time shows the yearly sawtooth: storage builds
through summer (injection) and drains through winter (withdrawal). Each peak
is roughly November (storage full, heading into winter); each trough is
roughly March (storage drained, end of heating season).

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(11, 5))

# plot storage over time. x: date, y: storage level
ax.plot(storage["Date"], storage["storage_bcf"], color="steelblue", linewidth=1)

ax.set_ylabel("Working gas in storage (Bcf)")
ax.set_xlabel("Year")
ax.set_title("US natural gas storage: the injection/withdrawal sawtooth (2010-2026)")
ax.grid(True, alpha=0.3)

plt.savefig("figures/storage_sawtooth.png", dpi=150, bbox_inches="tight")
plt.show()
```

### Storage vs price
If storage is the market's cushion, low storage should mean a tighter market
and firmer prices. I overlay storage against the monthly Henry Hub price to
look for that relationship.

```python
# storage is weekly, but price is monthly, aggregate storage to monthly to match. group by year AND month, average the weekly readings into one monthly number.
storage_monthly = storage.groupby(["year", "month"])["storage_bcf"].mean().reset_index()
storage_monthly.head()
```

```python
price_monthly = gas.groupby(["year", "month"])["price"].mean().reset_index()
combined = pd.merge(
    storage_monthly,
    price_monthly,
    on=["year", "month"],
    how="inner")

combined.head()
```

```python
# correlation between storage level and price.-1 = perfect inverse, 0 = none, +1 = perfect positive.
correlation = combined["storage_bcf"].corr(combined["price"])
print("Storage-price correlation:", round(correlation, 3))
```

The raw storage–price correlation is weak (-0.18), which was not expected. Analysts instead might watch
storage *relative to normal* for the time of year, I will build that: for each month, compare storage to the average storage for that same month across all years — the deviation is the real
tightness signal.

```python
monthly_norm = combined.groupby("month")["storage_bcf"].transform("mean")

combined["storage_vs_normal"] = combined["storage_bcf"] - monthly_norm

dev_corr = combined["storage_vs_normal"].corr(combined["price"])
print("Storage-deviation vs price correlation:", round(dev_corr, 3))
```

Correcting for seasonality, the correlation jumps from -0.18 to **-0.54** — a
clear inverse relationship. When storage is low *relative to its seasonal
normal*, the market is genuinely tight and prices are higher.

```python
fig, ax = plt.subplots(figsize=(8, 5.5))

ax.scatter(combined["storage_vs_normal"], combined["price"],
           alpha=0.5, color="steelblue", edgecolor="none")

ax.axvline(x=0, color="gray", linestyle="--", linewidth=1)
ax.set_xlabel("Storage vs seasonal normal (Bcf)  —  negative = tighter than usual")
ax.set_ylabel("Henry Hub price ($/MMBtu)")
ax.set_title("Tighter-than-normal storage lines up with higher prices (corr -0.54)")
ax.grid(True, alpha=0.3)

plt.show()
```

```python
# two charts side by side to compare raw vs seasonally-adjusted
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5.5))

# LEFT:raw storage level vs price
ax1.scatter(combined["storage_bcf"], combined["price"],
            alpha=0.5, color="indianred", edgecolor="none")
ax1.set_xlabel("Raw storage level (Bcf)")
ax1.set_ylabel("Henry Hub price ($/MMBtu)")
ax1.set_title("Raw storage vs price (corr -0.18)")
ax1.grid(True, alpha=0.3)

# RIGHT: storage vs seasonal normal
ax2.scatter(combined["storage_vs_normal"], combined["price"],
            alpha=0.5, color="steelblue", edgecolor="none")
ax2.axvline(x=0, color="gray", linestyle="--", linewidth=1)
ax2.set_xlabel("Storage vs seasonal normal (Bcf)")
ax2.set_ylabel("Henry Hub price ($/MMBtu)")
ax2.set_title("Seasonally-adjusted vs price (corr -0.54)")
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("figures/storage_price_comparison.png", dpi=150, bbox_inches="tight")
plt.show()
```

## What AI data-center demand does to the gas market
Analyst forecasts for AI-driven gas demand by 2030 span a wide range (sources
below). I model three scenarios and trace their effect through my storage-price relationship.
Sources for the demand range:
- Enverus (EIR, Dec 2025): ~2.1 Bcf/d confident, up to 4.1 if all-gas
- East Daley (2025): 4.2-6.1 Bcf/d base case
- RBC (May 2026): ~6.1 Bcf/d (range 6-7)
- S&P Global / EIA (2026): 3-6 Bcf/d base, up to 10-20 high; EIA high-demand
  scenario implies ~+$0.50/MMBtu to power-gen gas prices

Scenarios modeled: LOW 3 Bcf/d, MID 6 Bcf/d, HIGH 10 Bcf/d.

```python
ai_demand_low  = 3    
ai_demand_mid  = 6    #
ai_demand_high = 10   

# injection season is roughly April-October = ~214 days.
injection_days = 214

# total storage "lost" over an injection season, in Bcf, for each scenario
storage_hit_low  = ai_demand_low  * injection_days
storage_hit_mid  = ai_demand_mid  * injection_days
storage_hit_high = ai_demand_high * injection_days

print("Storage shortfall by end of injection season (Bcf):")
print("  low (3 Bcf/d): ", storage_hit_low)
print("  mid (6 Bcf/d): ", storage_hit_mid)
print("  high (10 Bcf/d):", storage_hit_high)
```

```python
import numpy as np

# Earlier I found storage-vs-normal correlates -0.54 with price. Get the SLOPE (dollars per Bcf) via a line of best fit, then convert each storage hit to $.
slope, intercept = np.polyfit(combined["storage_vs_normal"], combined["price"], 1)

price_impact_low  = -slope * storage_hit_low
price_impact_mid  = -slope * storage_hit_mid
price_impact_high = -slope * storage_hit_high

print("Slope:", round(slope, 5), "$/MMBtu per Bcf of storage deficit")
print()
print("Estimated Henry Hub price impact from AI gas demand:")
print("  low (3 Bcf/d):  +$", round(price_impact_low, 2), "/MMBtu")
print("  mid (6 Bcf/d):  +$", round(price_impact_mid, 2), "/MMBtu")
print("  high (10 Bcf/d): +$", round(price_impact_high, 2), "/MMBtu")
print()
print("EIA independent estimate (high-demand scenario): ~+$0.50/MMBtu")
```

```python
scenarios = ["Low\n(3 Bcf/d)", "Mid\n(6 Bcf/d)", "High\n(10 Bcf/d)"]
impacts = [price_impact_low, price_impact_mid, price_impact_high]

fig, ax = plt.subplots(figsize=(8.5, 5.5))

bars = ax.bar(scenarios, impacts, color=["#7fb069", "#f4a259", "#bc4b51"], edgecolor="black")

# EIA reference line
ax.axhline(y=0.50, color="steelblue", linestyle="--", linewidth=1.5)
ax.annotate("EIA supply-inclusive est. (~$0.50)", xy=(2.1, 0.6),
            color="steelblue", fontsize=9, ha="right")

for bar, val in zip(bars, impacts):
    ax.annotate(f"+${val:.2f}", xy=(bar.get_x() + bar.get_width()/2, val),
                xytext=(0, 4), textcoords="offset points", ha="center", fontsize=10)

ax.set_ylabel("Estimated Henry Hub price impact ($/MMBtu)")
ax.set_title("AI gas demand: modeled price impact (unmitigated) vs EIA estimate")
ax.grid(True, axis="y", alpha=0.3)

plt.savefig("figures/ai_gas_price_impact.png", dpi=150, bbox_inches="tight")
plt.show()
```
